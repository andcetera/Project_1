```python
import pandas as pd

testfile = 'disasters.csv'
df = pd.read_csv(testfile)
df.head()
```

```python
df.columns
```

```python
df['Disaster Type'].value_counts()
```

```python
climate_disasters = df.loc[(df['Disaster Subgroup'] == 'Climatological') | (df['Disaster Type'] == 'Storm') | (df['Disaster Type'] == 'Flood'), :]
climate_disasters.head()
```

```python
disaster_types = climate_disasters.groupby('Disaster Type')
disaster_types['Disaster Subtype'].value_counts()
```
